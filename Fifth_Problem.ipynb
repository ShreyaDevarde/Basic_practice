```python
def evenly_divisible(num):
    factors = []
    product = 0
    if num % 2 ==0:
        factors.append(2)
        num//=2
    i=3
    while i*i <=num:
        while i % 1 == 0:
            factors.append(i)
            num//=1
        i+=2
    if num >2:
        factors.append(num)
        
    return product
print(evenly_divisible(20))
        
```
