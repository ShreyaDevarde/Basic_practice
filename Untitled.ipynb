```python
num = int(input("enter the number:"))

if num <=1:
    print("Number is not prime enter greater than 1.")
else:
    i = 2
    is_prime = True
    while i<-num:
        if num % i==0:
            is_prime=False
            break
        i+=1
    if is_prime:
        print(f"{num} The Number is prime")
    else:
        print("Number is not prime")
        
```
