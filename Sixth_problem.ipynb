```python
import math
def square_diff(n):
    sum_square = 0
    square_sum = 0
    sum_s =0
    for i in range(1,n+1):
        sum_square += math.pow(i,2)
        sum_s +=i
    square_sum += math.pow(sum_s,2)
    difference = abs(sum_square - square_sum)
    return int(difference)
n = 100
print(square_diff(n))
        
```
