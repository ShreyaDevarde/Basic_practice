#### 17th Problem

```python
def number_to_words(n,include_and=True):
    if n == 1000:
        return "One Thousand"
    
    ones = ["", "One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", 
            "Eleven", "Twelve", "Thirteen", "Fourteen", "Fifteen", "Sixteen", "Seventeen", "Eighteen", "Nineteen"]
    
    tens = ["", "", "Twenty", "Thirty", "Forty", "Fifty", "Sixty", "Seventy", "Eighty", "Ninety"]
    
    result = ""

    # 1. Handle Hundreds
    if n >= 100:
        result += ones[n // 100] + " Hundred"
        n %= 100
        if n > 0:
            result += " " # Add space if there's more coming

    # 2. Handle Tens and Ones
    if n > 0:
        if n < 20:
            result += ones[n]
        else:
            ten_digit = n // 10
            one_digit = n % 10
            result += tens[ten_digit]
            if one_digit > 0:
                result += " " + ones[one_digit]

    return result

# Loop from 1 to 1000
for i in range(1, 1001):
    print(f"{number_to_words(i)}")
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


def is_truncatable(n):
    s = str(n)

    # left to right
    for i in range(len(s)):
        if not is_prime(int(s[i:])):
            return False

    # right to left
    for i in range(len(s)):
        if not is_prime(int(s[:len(s)-i])):
            return False

    return True


count = 0
total = 0
num = 11   # start from 2-digit numbers

while count < 11:
    if is_truncatable(num):
        total += num
        count += 1
    num += 1

print(total)
```

```python
import math

def count_distinct_prime_factors(n):
    factors = set()
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.add(d)
            n //= d
        d += 1
    if n > 1:
        factors.add(n)
    return len(factors)

def compute():
    i = 1
    while True:
        if all(count_distinct_prime_factors(i + j) == 4 for j in range(4)):
            return i
        i += 1

print(compute())
```
